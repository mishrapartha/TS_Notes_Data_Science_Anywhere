In general time series data is composed of 4 components

1. Trend
2. Cyclicality
3. Seasonality
4. Irregularity

**Long term general direction** is referred to as **Trend**

**Cyclicality** = **Patterns of highs and lows** through which data moves over time periods **usually of more than a year**

**Seasonality** = **Are shorter cycles**, which **usually occur in time periods of less than one year**

**Irregularity** = **Random and Noisy behaviour**, **rapid changes or 'bleeps' in the data**, which occur in even shorter time frames than seasonal effects
